```python
# Cellule 1 — Charger les librairies
import fitz  # pour lire les PDF
import os    # pour gérer les dossiers

print("Librairies chargées !")
```

```python

import os
os.chdir("..")  # remonter d'un dossier
pdf = fitz.open("data/raw/document.pdf")
print(f"{len(pdf)} pages trouvées")
```

```python
# Cellule 3 — Extraire toutes les pages en images JPG
os.makedirs("data/images", exist_ok=True)  # créer le dossier si inexistant

for i, page in enumerate(pdf):
    mat = fitz.Matrix(300/72, 300/72)        # 300 DPI = haute résolution
    pix = page.get_pixmap(matrix=mat)        # convertir la page en image
    pix.save(f"data/images/page_{i+1:03d}.jpg")  # sauvegarder
    print(f"Page {i+1}/34 extraite")
```

```python
# Cellule 4 — Vérifier le résultat
images = os.listdir("data/images")
print(f"{len(images)} images créées dans data/images/")
print(f"Première image : {images[0]}")
print(f"Dernière image : {images[-1]}")
```
